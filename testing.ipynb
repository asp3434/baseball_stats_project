```python
import requests
from bs4 import BeautifulSoup
import re
import time

testing_url = 'https://www.baseball-reference.com/teams/WSN/2019.shtml'

page = requests.get(testing_url)

soup = BeautifulSoup(page.content)

for this_soup in soup.find_all("td", {"class": "left", "data-stat": "player"}):
    try:
        print(this_soup.find("a")['href'])
    except:
        pass
```
